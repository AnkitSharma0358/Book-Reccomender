```python
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=";", error_bad_lines=False, encoding='latin-1')
```

```python
ratings.head(2)
```

```python
ratings.rename(columns={'User-ID':'user-id','Book-Rating':'ratings'}, inplace=True)
```

```python
ratings.rename(columns={'ratings':'rating'},inplace=True)
```

```python
x=ratings['user-id'].value_counts()>200
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user-id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_of_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_of_rating.rename(columns={'rating':'number of rating'}, inplace=True)
```

```python
number_of_rating
```

```python
final_rating=ratings_with_books.merge(number_of_rating, on='title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['number of rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user-id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user-id',index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestion=model.kneighbors(book_pivot.iloc[54, :].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
def reccommend_book(book_name):
    id_of_book=np.where(book_pivot.index==book_name)[0][0]
    distances, suggestion=model.kneighbors(book_pivot.iloc[id_of_book, :].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestion)):

            print(book_pivot.index[suggestion[i]])
```

```python
reccommend_book('Second Nature')
```
